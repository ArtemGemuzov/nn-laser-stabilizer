# RIN: Free-run и SIL

```python
import colorsys
from pathlib import Path

import matplotlib.pyplot as plt
import pandas as pd


def hsl_to_rgb_normalized(h, s, l):
    return colorsys.hls_to_rgb(h / 360, l / 100, s / 100)


BLUE_RGB = hsl_to_rgb_normalized(206, 73, 48)
GREEN_RGB = hsl_to_rgb_normalized(120, 72, 44)
RED_RGB = hsl_to_rgb_normalized(9, 98, 63)
PURPLE_RGB = hsl_to_rgb_normalized(279, 98, 76)
LIGHT_PURPLE_RGB = hsl_to_rgb_normalized(278, 100, 94)

GRAY_RGB = (123 / 255, 123 / 255, 123 / 255)

plt.rcParams.update(
    {
        "font.size": 24,
        "axes.titlesize": 24,
        "axes.labelsize": 24,
        "xtick.labelsize": 24,
        "ytick.labelsize": 24,
        "legend.fontsize": 24,
    }
)
```

```python
DATA_DIR = Path("..") / "data" / "Характеристики излучения"

rin_fr = pd.read_csv(
    DATA_DIR / "RIN_FR.csv",
    header=None,
    names=["frequency_hz", "rin_db_hz"],
)
rin_sil = pd.read_csv(
    DATA_DIR / "RIN_SIL.csv",
    header=None,
    names=["frequency_hz", "rin_db_hz"],
)
```

```python
fig, ax = plt.subplots(figsize=(14, 7))

ax.plot(rin_sil["frequency_hz"], rin_sil["rin_db_hz"], color=RED_RGB, linewidth=1.0, label="SIL")
ax.plot(rin_fr["frequency_hz"], rin_fr["rin_db_hz"], color=BLUE_RGB, linewidth=1.0, label="Free-run")

ax.set_xscale("log")

ax.set_xlabel("Frequency, Hz")
ax.set_ylabel("RIN, dBc/Hz")

ax.set_xlim(6.0, 5e4)
ax.set_ylim(-200, -60)

ax.grid(False)
ax.legend(loc="lower left", frameon=True)

plt.tight_layout()
plt.show()
```
